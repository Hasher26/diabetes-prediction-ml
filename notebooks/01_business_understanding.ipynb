## 2 · Background: Diabetes as a Public Health Problem

*Source: Burrows et al. (2017). CDC MMWR 66(43).*

### What the Paper Reports

Burrows et al. analysed the incidence of **End-Stage Renal Disease (ESRD)** attributed to diabetes in the United States and Puerto Rico from 2000 to 2014, using data from the U.S. Renal Data System (USRDS).

**Key findings (factual summary, no interpretation):**

- In 2014, approximately **120,000 people** in the US were newly diagnosed with ESRD; diabetes was the attributed cause in roughly **44%** of cases.
- ESRD incidence attributed to diabetes **declined** from 2000 to 2014, from approximately 162 to 122 per 100,000 adults with diagnosed diabetes.
- This decline was observed across racial/ethnic groups and age categories, though absolute rates varied substantially by subgroup.
- Puerto Rico showed different trends from the US mainland, with slower decline rates.

### Connection to BRFSS Data

The BRFSS (Behavioral Risk Factor Surveillance System) is the same data infrastructure that Burrows et al. reference as the source for diabetes prevalence denominators. Our dataset (2015) is a direct product of this surveillance system.

**Note:** BRFSS data is self-reported and population-level, while Burrows et al. use clinical registry data (USRDS). These are complementary but methodologically different data sources.

## 6 · Constraints & Known Limitations

### Known Dataset Limitations (factual – no team decision needed)

The following limitations are inherent to the BRFSS 2015 dataset and are non-negotiable:

| Limitation | Detail |
|------------|--------|
| **Self-reported data** | No clinical tests. Respondents self-declare diabetes status and health behaviours. |
| **Geographic scope** | USA + Puerto Rico only. Results may not generalise to other populations. |
| **Temporal scope** | Data from 2015 only. Health trends and diagnostic criteria may have changed. |
| **Target ambiguity** | `Diabetes_binary = 1` combines prediabetes and diabetes; the two are not distinguished. |
| **No Type 1 / Type 2 distinction** | The dataset does not differentiate between diabetes types. |
| **No longitudinal data** | Single cross-sectional survey; no follow-up information. |
