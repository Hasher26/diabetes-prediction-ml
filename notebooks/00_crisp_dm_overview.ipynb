## What is CRISP-DM?

CRISP-DM (Cross-Industry Standard Process for Data Mining) is an **iterative, non-linear process model** for data mining and machine learning projects. It was introduced by a consortium of European companies and formalised in:

> Wirth, R., & Hipp, J. (2000). *CRISP-DM: Towards a standard process model for data mining.* Proceedings of the 4th International Conference on the Practical Applications of Knowledge Discovery and Data Mining.

### The 6 Phases

| # | Phase | Description |
|---|-------|-------------|
| 1 | **Business Understanding** | Define the problem from a business perspective. Set objectives, success criteria, and constraints before touching the data. |
| 2 | **Data Understanding** | Collect, explore, and assess the data. Identify quality issues and form initial hypotheses. |
| 3 | **Data Preparation** | Clean, transform, and engineer features. Construct the final modelling dataset. |
| 4 | **Modeling** | Select and apply modelling techniques. Tune hyperparameters. |
| 5 | **Evaluation** | Assess whether the model meets the business objectives. Determine whether to deploy or iterate further. |
| 6 | **Deployment** | Plan and document how results will be used, communicated, or integrated into a system. |

### CRISP-DM is a Cycle, Not a Linear Sequence

A defining property of CRISP-DM is that it is **cyclical and iterative**. The process does not end after Deployment:

- Evaluation may reveal that **Business Understanding was incomplete** → return to Phase 1  
- Modelling may expose unexpected **data quality issues** → return to Phase 3  
- Feedback after Deployment may demand a **revised model** → re-enter at Phase 2 or 3  

This iterative design makes CRISP-DM realistic for applied projects where requirements and data realities evolve over time.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

PLOTS_DIR = os.path.join('..', 'results', 'plots')
os.makedirs(PLOTS_DIR, exist_ok=True)

phases = [
    '1. Business\nUnderstanding',
    '2. Data\nUnderstanding',
    '3. Data\nPreparation',
    '4. Modeling',
    '5. Evaluation',
    '6. Deployment',
]
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F', '#EDC948']

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.axis('off')

outer_r = 1.0
inner_r = 0.42
n = len(phases)
seg = 360 / n

for i, (phase, color) in enumerate(zip(phases, colors)):
    ca = 90 - i * seg
    wedge = mpatches.Wedge(
        center=(0, 0), r=outer_r,
        theta1=ca - seg / 2, theta2=ca + seg / 2,
        width=outer_r - inner_r,
        facecolor=color, edgecolor='white', linewidth=2.5, alpha=0.92, zorder=3
    )
    ax.add_patch(wedge)
    r_label = (outer_r + inner_r) / 2
    ax.text(
        r_label * np.cos(np.radians(ca)),
        r_label * np.sin(np.radians(ca)),
        phase, ha='center', va='center',
        fontsize=9.5, fontweight='bold', color='white', zorder=4
    )

# White center circle
ax.add_patch(plt.Circle((0, 0), inner_r, color='white', zorder=4))
ax.text(0, 0.08, 'CRISP-DM', ha='center', va='center',
        fontsize=13, fontweight='bold', color='#333333', zorder=6)
ax.text(0, -0.12, 'Iterative\nProcess', ha='center', va='center',
        fontsize=9.5, color='#777777', zorder=6)

# Clockwise arrow ring around the outside
arrow_r = outer_r + 0.16
t = np.linspace(np.radians(100), np.radians(100 - 348), 300)
ax.plot(arrow_r * np.cos(t), arrow_r * np.sin(t), color='#aaaaaa', lw=2, zorder=2)
ax.annotate('',
            xy=(arrow_r * np.cos(t[-1]), arrow_r * np.sin(t[-1])),
            xytext=(arrow_r * np.cos(t[-15]), arrow_r * np.sin(t[-15])),
            arrowprops=dict(arrowstyle='->', color='#aaaaaa', lw=2), zorder=2)

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_title('CRISP-DM Process Model  (Wirth & Hipp, 2000)',
             fontsize=13, fontweight='bold', pad=15, color='#444444')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, '00_crisp_dm_diagram.png'), dpi=150, bbox_inches='tight')
plt.show()
print('Saved to results/plots/00_crisp_dm_diagram.png')
```

## Why CRISP-DM for This Project?

CRISP-DM is particularly well-suited for healthcare machine learning projects. Four concrete reasons:

### 1. Iterative Approach Allows Mid-Course Corrections
Medical datasets regularly surface surprises during exploration: class imbalance, unexpected feature interactions, or domain-specific data quality issues. CRISP-DM's cyclical structure ensures these discoveries can feed back into earlier phases without abandoning the overall project structure.

### 2. Business Understanding First Prevents Irrelevant Models
By defining the clinical objective *before* any modelling, we avoid optimising for the wrong metric. For diabetes prediction, this distinction is critical: a model that maximises accuracy by always predicting "No Diabetes" achieves ~86% accuracy but is clinically useless. The decision to prioritise **Recall** or **AUC-ROC** belongs in Phase 1, not Phase 4.

### 3. Separating Data Preparation from Modeling Prevents Data Leakage
Explicitly separating Phase 3 and Phase 4 enforces the discipline of fitting all preprocessing steps (scaling, SMOTE, encoders) **only on training data** before applying them to the test set. This prevents optimistic bias — one of the most common failure modes in ML projects.

### 4. Evaluation Before Deployment Ensures Clinical Relevance
Phase 5 asks: *does this model meet the original business objective?* For a public health prediction tool, that means not just statistical performance but also interpretability, fairness across demographic subgroups, and probability calibration. This checkpoint prevents premature deployment of technically correct but clinically misleading models.

## Dataset Context

| Property | Detail |
|----------|--------|
| **Source** | CDC Behavioral Risk Factor Surveillance System (BRFSS), 2015 |
| **Repository** | UCI ML Repository, ID #891 |
| **Geography** | USA + Puerto Rico (telephone survey) |
| **Samples** | ~253,680 |
| **Features** | 21 (14 binary, 4 ordinal, 3 numeric) |
| **Missing values** | None |
| **Target variable** | `Diabetes_binary` |
| **Class distribution** | ~86% No Diabetes / ~14% Prediabetes or Diabetes |
| **Class imbalance ratio** | ~6:1 (majority:minority) |

### Target Variable

As specified by the course supervisor (Prof. Dr. Kevin Bauer):

- `0` → **No diabetes**
- `1` → **Prediabetes or diabetes** (combined into one positive class)

This formulation frames the task as **early risk detection** rather than precise clinical diagnosis.

### Key Limitation

BRFSS is a **self-reported** survey. Respondents self-declare their diabetes status and lifestyle behaviours without clinical verification. This introduces potential reporting bias and limits the generalisability of any learned model to populations that self-report health data similarly.

## Reference: The Public Health Motivation

### Burrows et al. (2017)

> Burrows, N. R., Hora, I., Geiss, L. S., Gregg, E. W., & Albright, A. (2017). Incidence of End-Stage Renal Disease Attributed to Diabetes Among Persons with Diagnosed Diabetes — United States and Puerto Rico, 2000–2014. *MMWR Morbidity and Mortality Weekly Report*, 66(43), 1165–1170.

### Connection to Our Project

Burrows et al. document the trajectory from **unmanaged diabetes → End-Stage Renal Disease (ESRD)**: a severe, costly, and largely preventable complication. Their key finding — that ESRD incidence *declined* over 2000–2014, likely due to improved diabetes management — demonstrates that early intervention is effective.

This motivates our modelling goal:

| Without Detection | With Early Detection (our model) |
|-------------------|----------------------------------|
| Diabetes goes undiagnosed | High-risk individuals flagged early |
| No treatment initiated | Medical referral and lifestyle intervention |
| Progression to ESRD and other complications | Reduced complication rates |

A model trained on BRFSS survey data could, in principle, support **population-level risk screening** — flagging high-risk individuals before clinical tests are administered. This is the deployment vision that frames our entire modelling effort.
